### Define distribution

```python
from sampler import Sampler, sample_autocorrelation

N = 4
moments = {
    (0,): -0.3,
    (1,): -0.35,
    (2,): -0.5,
    (3,): -0.11,
    (0, 1): 0.2,
    (2, 3): 0.5
}
sampler = Sampler(N, moments, sample_interval=5)
```

### Check autocorrelation

```python
import matplotlib.pyplot as plt
samples_autocor = sampler.sample_multiple(10000)
autocor = sample_autocorrelation(N, samples_autocor)
plt.plot(autocor)
plt.xlim([len(autocor)/2-10, len(autocor)/2+10])
```

### Sample and compare to exact distribution

```python
from collections import Counter

num_samples = 10000
samples = sampler.sample_multiple(num_samples)
sample_freq = Counter(samples)
```

```python
import numpy as np
import itertools as it

all_xs = list(it.product(*(N * [[0, 1]])))
p_unnorm = lambda x: np.exp(np.sum([
    c * np.product([x[i] for i in m])
    for m, c in moments.items()
]))
Z = np.sum([p_unnorm(x) for x in all_xs])
p = lambda x: p_unnorm(x) / Z
```

```python
import matplotlib.pyplot as plt
exact_p = [p(x) for x in all_xs]
sampled_p = [sample_freq.get(tuple([i for i in range(N) if x[i] == 1]), 0) / num_samples for x in all_xs]
plt.scatter(exact_p, sampled_p)
m = min(exact_p + sampled_p)
M = max(exact_p + sampled_p)
plt.plot([m, M], [m, M], 'r--')
plt.xlabel('exact probability')
plt.ylabel('empirical probability from sampler')
```
